# Line Charts

```python
# Import critical libraries
import matplotlib.pyplot as plt
import pandas as pd
```

## WHO time series data for plot

```python
who_time_series = pd.read_csv('data/WHO_time_series.csv')
who_time_series.head()
```

```python
who_time_series.info()
```

```python
who_time_series['Country'].nunique()
```

```python
who_time_series.head()
```

```python
who_time_series['Date_reported'] = pd.to_datetime(who_time_series['Date_reported'])
display(who_time_series.head())
display(who_time_series.info())
display(who_time_series.isnull().sum())
```

```python
italy = who_time_series[who_time_series['Country'] == 'Italy']
india = who_time_series[who_time_series['Country'] == 'India']
china = who_time_series[who_time_series['Country'] == 'China']
poland = who_time_series[who_time_series['Country'] == 'Poland']

plt.plot(poland['Date_reported'], poland['Cumulative_cases'], 'g', label = 'Poland')
plt.plot(italy['Date_reported'], italy['Cumulative_cases'], 'y', label = 'Italy')
plt.plot(india['Date_reported'], india['Cumulative_cases'], 'r', label = 'Poland')

plt.title('Cumulative Reported Cases')
plt.xlabel('Date')
plt.ylabel('Number Of Cases')
plt.legend()
plt.show()
```

```python
# Define a function to create plots

def plot_cumulative_cases(country_name):
    country = who_time_series[who_time_series['Country'] == country_name]
    plt.plot(country['Date_reported'], country['Cumulative_cases'])
    plt.title('{}: Cumulative Reported Cases'.format(country_name))
    plt.xlabel('Date')
    plt.ylabel('Number of Cases')
    plt.show()
    
```

```python
plot_cumulative_cases('Norway')
plot_cumulative_cases('India')
```

```python
france = who_time_series[who_time_series['Country'] == 'France']
uk = who_time_series[who_time_series['Country'] == 'The United Kingdom']
italy = who_time_series[who_time_series['Country'] == 'Italy']

countries = [france, uk, italy]
labels = ['France', 'The UK', 'Italy']

for c, l in zip(countries, labels):
    plt.plot(c['Date_reported'], c['Cumulative_cases'], label = l)

plt.legend()
plt.show()
```

# Scatter Plots and Correlations

```python
bike_sharing = pd.read_csv('data/day.csv')
display(bike_sharing.head())
display(bike_sharing.info())
```

```python
bike_sharing.isnull().sum()
```

```python
# Start exploration - plot line chart of counts vs date
x = bike_sharing['dteday']
y = bike_sharing['cnt']

plt.plot(x,y)
plt.show()
```

```python
# dteday is an object - convert to date
bike_sharing['dteday'] = pd.to_datetime(bike_sharing['dteday'])
```

```python
x = bike_sharing['dteday']
y = bike_sharing['cnt']

plt.plot(x,y)
plt.xticks(rotation = 45)
plt.show()
```

```python
x = bike_sharing['dteday']
y1 = bike_sharing['casual']
y2 = bike_sharing['registered']

plt.plot(x,y1, label='Casual')
plt.plot(x,y2, label='Registered')

plt.xticks(rotation = 30)
plt.ylabel('Bike Rented')
plt.xlabel('Date')

plt.title('Bikes Rented: Casual vs. Registered')
plt.legend()

plt.show()

```

```python
'''
- Plot a line graph with the dteday column on the x-axis and the temp column on the y-axis.
- Rotate the x-ticks to an angle of 45 degrees using plt.xticks().
- Do you see any similarity in how the temperature and the number of bikes rented change?
'''

x = bike_sharing['dteday']
y1 = bike_sharing['temp']
y2 = bike_sharing['cnt']

plt.plot(x,y1, label = 'Temperature')
plt.plot(x,y2, label = 'Bike Rentals')

plt.xticks(rotation = 45)
plt.title('Temprature overtime')
plt.legend()
plt.show()
```

Above, we plotted `cnt` and temp against `dteday`. To visualize the relationship between cnt and temp, it's better if we plot one against the other, and leave out the dteday column. We can do this with a **scatter plot** (also known as scatter graph or scatter chart).

```python
# Scatter plots
x = bike_sharing['temp']
y = bike_sharing['cnt']

plt.scatter(x,y)
plt.title('Temperature vs Rentals')
plt.xlabel('Temp')
plt.ylabel('Rentals')

plt.show()
display(x.corr(y))
```

```python
# Taking a more granular approach

bike_sharing['weekday'].value_counts().sort_index()
```

```python
# Calculate the average grouped by weekdays

grouped_df = bike_sharing.groupby('weekday')
weekday_avgs = round(grouped_df[['casual','registered']].mean())
weekday_avgs
```

```python
# Create a bar chart for casual by weekdays 
plt.bar(x=weekday_avgs.index, height=weekday_avgs['casual'])
plt.title('Average rentals by weekday - Casual Users')
plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6],
          labels=['Sunday', 'Monday', 'Tuesday', 'Wednesday',
                 'Thursday', 'Friday', 'Saturday'], rotation = 30)
plt.show()
```

```python
# Create a bar chart for registered by weekdays 
plt.bar(x=weekday_avgs.index, height=weekday_avgs['registered'])
plt.title('Average rentals by weekday - Registered Users')
plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6],
          labels=['Sunday', 'Monday', 'Tuesday', 'Wednesday',
                 'Thursday', 'Friday', 'Saturday'], rotation = 30)
plt.show()
```

## Grouped Frequency Tables and Histograms

```python
# Creating frequency tables for continuous variables
rental_dist = bike_sharing['cnt'].value_counts(bins = 10).sort_index()
rental_dist
```

```python
plt.figure(figsize=(10,5))
plt.hist(bike_sharing['cnt'], bins=10)
plt.title('Bike rental distribution')
plt.xlabel('Bike Rentals')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(bike_sharing['registered'], bins=10)
plt.title('Registered rental distribution')
plt.xlabel('Bike Rentals - Registsred')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10,5))
plt.hist(bike_sharing['casual'], bins=10)
plt.title('Casual rental distribution')
plt.xlabel('Bike Rentals - Casual')
plt.ylabel('Frequency')
plt.show()
```
