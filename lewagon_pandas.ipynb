## Building block - series

```python
import pandas as pd
```

```python
# 2 Ways of generating series - use pd.Series method in both the cases
# use two lists one each for data and index 
data=[1,2,3,'four']
index = ['i1','i2','i3','i4']
s = pd.Series(data = data, index = index)
s
```

```python
# Second method is to use a dictionary
data = {
    'i1':1,
    'i2':2,
    'i3':3,
    'i4':'four'
}
s = pd.Series(data)
s
```

## Grouping and Aggregate functions

There are 3 steps for grouping
1. **Split:** Split and group the data into subsets
2. **Apply:** An aggregare function like sum, mean etc
3. **Combine** The result of two previous operation and create a new dataset 

```python
# Which is the most populous region in the world
regions = countries_df.groupby('Region')
```

```python
# Select columns are apply aggregate functions
region_population_df = regions[['Population', 'Area (sq. mi.)' ]].sum().sort_values(by='Population', ascending=False)
```
